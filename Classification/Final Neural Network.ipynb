```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mglearn

df = pd.read_csv("processed_classification.csv")
df = df[['Sex_M', 'Sex_F', 'Age', 'Time', 'Nbr_Warts', 'Type_1',
       'Type_2', 'Type_3', 'Area', 'Induration_diameter', 'Result']]

df
```

```python
y = df['Result']

X_DataFrame = df[['Sex_M', 'Sex_F', 'Age', 'Time', 'Nbr_Warts', 'Type_1',
       'Type_2', 'Type_3', 'Area', 'Induration_diameter']]

X = X_DataFrame.to_numpy()
```

# Split the Dataset intoTrain and Test sets

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)

print (" train X = ", X_train.shape, " --- train y = ", y_train.shape)

print (" test X = ", X_test.shape, " --- test Y = ", y_test.shape)
```

# Testing hyperparameters for the classifier

```python
from sklearn.neural_network import MLPClassifier
from sklearn import metrics

for i in range(1, 4):
    for j in range(1, 10):
        for k in ['adam', 'sgd', 'lbfgs']:          
            for l in np.linspace(0.0001, 0.05, 10):
                clf = MLPClassifier(solver = k,  alpha = l, hidden_layer_sizes=(i, j), random_state = 1, max_iter = 500)
                clf.fit(X_train, y_train)
                y_pred = clf.predict(X_test)
                if (metrics.accuracy_score(y_pred,y_test) >= 0.89):
                    print('The accuracy of the Multi-layer Perceptron is :{} for {} layers, {} neurons, {} alpha, and {} solver'.format(metrics.accuracy_score(y_pred,y_test), i, j, l, k))
```

# Trying to optimizing hyperparameters using GridSerachCV

```python
#We try to optimize the classifier hyperparamaters using GridsearchCV algorithm (other than the number and sizes of hidden layers)
#It does the process of performing hyperparameter tuning in order to determine the optimal values for a given model
#we pass predefined values for hyperparameters to the GridSearchCV function by defining a dictionary in which we mention a particular hyperparameter along with the values it can take

parameter_space = {
    'hidden_layer_sizes': (2, 9),
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam', 'lbfgs'],
    'alpha': np.linspace(0.0001, 0.05, 100),
    'learning_rate': ['constant','adaptive'],
}
```

```python
#Initializing the GridsearchCV

from sklearn.model_selection import GridSearchCV

mlp_grid = MLPClassifier(max_iter=(200))

clf_grid = GridSearchCV(mlp_grid, parameter_space, n_jobs=-1, cv=5)
clf_grid.fit(X_train, y_train)
```

```python
#To report different results based on the set of parameters 

means = clf_grid.cv_results_['mean_test_score']
stds = clf_grid.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf_grid.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
```

```python
#To get the best parameters

print('Best parameters found:\n', clf_grid.best_params_)
```

```python
#To test the accuracy of the best parameters

clf_test = MLPClassifier(solver='lbfgs', activation = 'tanh', alpha = 0.002116161616161616, hidden_layer_sizes=(2, 9), learning_rate = 'adaptive', max_iter=200, random_state=1)
clf_test.fit(X_train, y_train)
y_pred_test = clf_test.predict(X_test)

metrics.accuracy_score(y_pred_test,y_test)
```

```python
from sklearn.metrics import classification_report

y_grid, y_pred_grid = y_test , clf_grid.predict(X_test)

print('Results on the test set:')
print(classification_report(y_grid, y_pred_grid))
```
